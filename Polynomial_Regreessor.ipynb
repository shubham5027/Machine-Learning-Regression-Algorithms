```python
px.scatter(df,x='Position',y='Salary')
```

```python
X=df.drop(['Salary','Position'],axis=1)
Y=df[['Salary']]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
X_train.head()
```

Polynomial is on sort of linear regression algorithm , as we just appply polynomial features as a preprocessing techinique with degre as a hyperparameter and after preprocessing fit to linearRegression Model

```python
trf2=PolynomialFeatures(degree=5) #degree=2   r2_score=70.324
                                  #degree=3   r2_score=0.8720
                                  #degree=4   r2_score=0.9635
                                  #degree=5   r2_score=0.9943 (Excellent)
```

```python
trf3=LinearRegression()
```

```python
pipe=make_pipeline(trf2,trf3)
```

```python
pipe=pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
r2=r2_score(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
```

```python
print(r2,mse,mae)
```
